```python
import pandas as pd

# 读取原始数据集
df = pd.read_csv('database.csv')

# 检查数据集中是否存在缺失值，确保基础数据的完整性
print("缺失值检查结果：")
print(df.isnull().sum())

# 将文本格式的日期转换为 pandas 标准日期对象
df['Date'] = pd.to_datetime(df['Date'])

# 从日期中提取月份和星期，为后续的周期性趋势分析做准备
df['Month'] = df['Date'].dt.month
df['DayOfWeek'] = df['Date'].dt.day_name()

# 将文本格式的时间转换为时间对象，并提取出具体的小时数
# 这对分析超市每天的客流高峰时段非常关键
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M').dt.hour

# 模拟运营报表视角：按商品类别计算核心财务与满意度指标
category_summary = df.groupby('Product line').agg(
    Total_Revenue=('Total', 'sum'),
    Total_Gross_Income=('gross income', 'sum'),
    Average_Rating=('Rating', 'mean')
).reset_index()

# 保留两位小数，使得最终输出的表格更加整洁
category_summary['Total_Revenue'] = category_summary['Total_Revenue'].round(2)
category_summary['Total_Gross_Income'] = category_summary['Total_Gross_Income'].round(2)
category_summary['Average_Rating'] = category_summary['Average_Rating'].round(2)

print("\n各商品类别的销售额、毛利润与平均顾客评分：")
display(category_summary)

# 模拟顾客洞察视角：对比会员与普通顾客的消费习惯
customer_summary = df.groupby('Customer type').agg(
    Average_Spend_Per_Transaction=('Total', 'mean'),
    Total_Transactions=('Invoice ID', 'count')
).reset_index()

customer_summary['Average_Spend_Per_Transaction'] = customer_summary['Average_Spend_Per_Transaction'].round(2)

print("\n会员与普通顾客的平均客单价与交易总笔数对比：")
display(customer_summary)
```
